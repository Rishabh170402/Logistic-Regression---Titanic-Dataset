```python
import pandas as pd
import numpy as np
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
df= pd.read_csv("traine.csv")
X=df.drop(['Survived','Cabin','Age','Name','Embarked','Ticket'], axis=1)
Y=df['Survived']

X.head()
```

```python
def f(s):
    if s=="male":
        return 0
    else:
        return 1
X["Gender"]=X.Sex.apply(f)
X.head()
```

```python
del X["Sex"]
```

```python
X.head()
```

```python
clf.fit(X,Y)
```

```python
pf= pd.read_csv("teste.csv")
Output=pf.drop(['Cabin','Age','Name','Embarked','Ticket'], axis=1)
def f(s):
    if s=="male":
        return 0
    else:
        return 1
Output["Gender"]=Output.Sex.apply(f)
del Output["Sex"]
Output.head()
```
